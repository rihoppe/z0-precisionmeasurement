```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
Types = {b"ELECTRON":"g",b"MUON":"b",b"TAUS":"c",b"HADRON":"r"}
Header = ["Run","Event","PCharged","NCharged","E_ECal","E_HCal","Type"]
Data = np.genfromtxt('./GropeData/All.dat',skip_header=1,names=Header,dtype=[('Run', '<i8'), ('Event', '<i8'), ('PCharged', 'f8'), ('NCharged', '<i8'), ('E_ECal', '<f8'), ('E_HCal', '<f8'), ('Type', 'S8')])
```

```python
fig = plt.figure()
for i in Types.keys():
    plt.hist(Data[Data["Type"]==i]["NCharged"], bins=15, histtype="step", color=Types[i], label=i, range=(0,70))

fig.legend(bbox_to_anchor=(0.4,0.38,0.5,0.5))
fig.patch.set_facecolor("white")
plt.xlabel("Number Tracks")
plt.ylabel("Events")
fig.show()
```

```python
indx_eff = Data[np.where(Data["NCharged"]>=15)]
eff = 100*len(np.where(indx_eff["Type"]==b"HADRON")[0])/len(np.where(Data["Type"]==b"HADRON")[0])
signal_noise = len(np.where(indx_eff["Type"]==b"HADRON")[0])/len(indx_eff["Type"])
print("Efficiency = " + "%.1f" % eff + "%" )
print("S/N = " + "%f" % signal_noise)
```
